```python
import hashlib

nounce = 1
txA = 'Hello'.encode()
txB = 'How are you?'.encode()
txC = 'This is Thursday'.encode()
txD = 'Happy new Year'.encode()
```

```python
hashA = hashlib.sha256(txA)
hashB = hashlib.sha256(txB)
```

```python
hashA.hexdigest() #hashA의 hex값 구하기
```

```python
for i in range(0,hashA.digest_size*2,2):
    print(hashA.hexdigest()[i:i+2]) #2자리씩 끊어서 출력
```

```python
hashAswap="".join(reversed([hashA.hexdigest()[i:i+2] for i in range(0,hashA.digest_size*2,2)]))#byte swqp => reversed함수
```

```python
hashBswap="".join(reversed([hashB.hexdigest()[i:i+2] for i in range(0,hashB.digest_size*2,2)]))
```

```python
print("hashAswap: ", hashAswap, "\nhanshBswap: ", hashBswap)
```

```python
hashAB=hashAswap+hashBswap 
```

```python
doubleHashAB=hashlib.sha256(hashAB.encode()).hexdigest() #더블해싱
print(doubleHashAB) #거래1,2의 해쉬값
```

```python
import hashlib
ntry=1 #시도 (여기서는 필요x)
found=False #loop를 마칠지(while문에 들어가기 위함)
blockNumber=54 #임의의 값  hex값
NOUNCE=0 #임의값
data='Hello'
previousHash='5d7c7ba21cbbcd75d14800b100252d5b428e5b1213d27c385bc141ca6b47989e'
while found==False:
    z = str(blockNumber)+str(NOUNCE)+data+previousHash
    guessHash = hashlib.sha256(z.encode('utf-8')).hexdigest()
    if guessHash[:4]=='00000':
        found=True
    NOUNCE+=1
    if(NOUNCE%10000000==0):
        print("NOUNCE: ",NOUNCE, guessHash)
print(guessHash)
print(NOUNCE)

```
